```python
import seaborn as sns

sns.set(style="whitegrid")
f=sns.load_dataset("fmri")

sns.scatterplot(x="timepoint",y="signal",data=f)
```

```python
sns.scatterplot(x="timepoint",y="signal",data=f,hue="region",style="event")
```

```python
p=sns.load_dataset("tips")
sns.scatterplot(x="day",y="tip",data=p)
```

```python
sns.scatterplot(x="day",y="tip",data=p,marker="d")
```

```python
sns.scatterplot(x="day",y="tip",data=p,hue="time")
```

```python
sns.scatterplot(x="day",y="tip",data=p,hue="day")
```

```python
sns.scatterplot(x="day",y="tip",data=p,hue="time",style="time")
```

```python
sns.scatterplot(x="day",y="tip",data=p,hue="time",palette="pastel")
```

```python
sns.scatterplot(x="day",y="tip",data=p,hue="day",sizes=(30,200),legend="brief")
```
