```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
a=[1,2,3,4,5]
b=[2,3,4,5,6]
sns.lineplot(x=a,y=b)
```

```python
a=[1,2,3,4,5]
b=[2,3,4,5,6]
s=pd.DataFrame({'a':a,'b':b})
print(s)

sns.lineplot(x='a',y='b',data=s)
```

```python
df= sns.load_dataset('penguins')
print(df)   


sns.lineplot(x='flipper_length_mm',y='body_mass_g',data=df)
```

```python
 
sns.lineplot(x='flipper_length_mm',y='body_mass_g',data=df,hue='species')
```

```python
sns.lineplot(x='flipper_length_mm',y='body_mass_g',data=df,hue='species',style='sex'   )
```

```python
sns.lineplot(x='flipper_length_mm',y='body_mass_g',
             data=df,hue='species',style='sex',palette='Set2'   )  
```

```python
sns.lineplot(x='flipper_length_mm',y='body_mass_g',
             data=df,hue='species',style='sex',palette='Set2',markers=["o","s"]   )
```

```python
df2=sns.load_dataset('iris')

print(df2)

sns.lineplot(x='sepal_length',y='sepal_width',
             data=df2,hue='species',style='species',palette='Set2',markers=["o","s","d"]   )
```

```python
df3=sns.load_dataset('tips')
print(df3)
sns.lineplot(x='total_bill',y='tip',data=df3,hue='day',style='day',palette='Set2',markers=["o","s","d","^"]   )
```
