```python
import seaborn as sns
e=sns.load_dataset("exercise")
sns.catplot(x="time",y="pulse",hue="kind",data=e)

#default gives scatter
```

```python
sns.catplot(x="time",hue="kind",data=e,kind="count")
```

```python
sns.catplot(x="time",y="pulse",hue="kind",data=e,kind="bar")
```

```python
sns.catplot(x="time",y="pulse",hue="kind",data=e,kind="violin")
```

```python
sns.catplot(x="time",y="pulse",hue="kind",data=e,col="diet")
```

```python
sns.catplot(x="time",y="pulse",hue="kind",data=e,kind="box")
```
