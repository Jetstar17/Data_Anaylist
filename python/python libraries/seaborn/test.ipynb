```python
import seaborn as sns
 
 
tips= sns.load_dataset("tips")
tips
```

```python
sns.pairplot(data=tips,hue="sex")
```

```python
sns.lineplot(x="total_bill",y="tip",hue="sex",data=tips,style="smoker")
```

```python
sns.barplot(x="total_bill",y="tip",data=tips)
```

```python
sns.countplot(y="tip",hue="sex",data=tips)
```

```python
sns.displot(x="total_bill",hue="sex",data=tips,kde=True)
```

```python
sns.lmplot(x="total_bill",y="tip",hue="sex",data=tips,col="smoker",row="sex")
```

```python
sns.violinplot(x="total_bill",y="tip",hue="sex",data=tips)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="sex",data=tips,style="sex",markers=("o","d"))
```

```python
sns.catplot(x="total_bill",y="tip",hue="sex",data=tips,kind="bar")
```

```python
sns.histplot(x="total_bill",y="tip",hue="sex",data=tips,element="poly")
```

```python
sns.swarmplot(x="total_bill",y="tip",hue="sex",data=tips)
```

```python
sns.kdeplot(x="total_bill",color="green",data=tips)
```
