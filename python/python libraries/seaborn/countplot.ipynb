```python
import seaborn as sns
p=sns.load_dataset("tips")
sns.countplot(x="sex",data=p)
```

```python
import seaborn as sns
p=sns.load_dataset("tips")
sns.countplot(x="sex",hue="smoker",data=p)
```

```python
sns.countplot(x="sex",hue="smoker",data=p,palette="Set2")
```

```python
sns.countplot(x="sex",hue="smoker",data=p,color="salmon")
```
