```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
ru=np.random.RandomState(10)

d=ru.normal(size=100)
sns.histplot(d,kde=True,color="m")
```

```python
sns.set(rc={"figure.figsize":(3,4)})
x=np.random.randn(10) 
ax=sns.histplot(x)
```

```python
from pylab import *
sns.set(rc={"figure.figsize" :(5,9)})
x=np.random.randn(10)
plt.subplot(2,2,1)
sns.distplot(x)
plt.subplot(2,2,2)
sns.distplot(x, rug=True,hist=False)
plt.subplot(2,2,3)
sns.distplot(x,vertical=True)
plt.subplot(2,2,4)
sns.kdeplot(x,shade=True,color="red")
```

```python
t=sns.load_dataset("titanic")
age1=t["age"].dropna()
sns.displot(age1,bins=30,kde=True)
```
