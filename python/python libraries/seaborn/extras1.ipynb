```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

plt.plot(iris["sepal_length"], iris["sepal_width"])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Sepal Length vs Width")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

sns.lineplot(data=penguins,
             x="bill_length_mm",
             y="flipper_length_mm",
             hue="sex")

plt.title("Penguin Growth Analysis")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
age = titanic["age"].dropna()

sns.histplot(age, bins=20, kde=False)

plt.title("Titanic Age Distribution")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
age = titanic["age"].dropna()

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.histplot(age, kde=True)
plt.title("Normal Distribution")

plt.subplot(1,2,2)
sns.histplot(age, kde=True)
sns.rugplot(age)
plt.title("Distribution with Rug")

plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

a=np.random.randn(100)
sns.set(rc={"figure.figsize":(8,8)})

plt.subplot(2,2,1)
sns.distplot(a,kde=True)

plt.subplot(2,2,2)
sns.distplot(a,kde=True,hist=False)

plt.subplot(2,2,3)
sns.distplot(a,vertical=True)

plt.subplot(2,2,4)
sns.distplot(a,kde=True,rug=True,color="red")

```
