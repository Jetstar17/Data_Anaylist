```python
import seaborn as sns

p=sns.load_dataset("tips")
sns.lmplot(x="total_bill",y="tip",col="day",hue="smoker",data=p)
```

```python
sns.lmplot(x="total_bill",y="tip",col="day",hue="smoker",data=p,col_wrap=2)
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=p,line_kws={"color":"red","lw":5,"ls":"--"})
#linw kws add line style line width color etcc
```

```python
sns.lmplot(x="total_bill",y="tip",hue="day",data=p,hue_order=["Sun","Sat","Fri","Thur"])
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=p,height=7,aspect=1.2)
# height is the size in inches,  and aspect is the width-to-height  ratio. Use this to prevent your plot from looking "squeezed."
```

```python
sns.lmplot(x="total_bill",y="tip",data=p,robust=True)
# Parameter robust:  If your data has "Outliers" (strange  points far from the rest), this te ls Seaborn to ignore them slightly so the line fits the majority of the data better.
```

```python
a=sns.load_dataset("anscombe")
sns.lmplot(x= "x" , y= "y" , data=a.query( "dataset  == 'II'" ), logx= True )
# Parameter logx:  Estimates a linear regression after  transforming the X variable with a logarithm. 
```

```python
tips=sns.load_dataset("tips")
z=sns.lmplot(x= "total_bill" , y= "tip" ,col="day", data=tips)
z.set_titles("{col_name}Analysis")
```

```python
custom_colors={"Male":"navy","Female":"pink"}
sns.lmplot(x= "total_bill" , y= "tip" , hue= "sex" , palette=custom_colors,  data=tips) 
```
