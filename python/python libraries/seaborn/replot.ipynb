```python
import seaborn as sns

t=sns.load_dataset("titanic")
sns.relplot(data=t,x="age",y="fare")
```

```python
sns.relplot(data=t,x="age",y="fare",hue="class")
```

```python
sns.relplot(data=t,x="age",y="fare",hue="class",col="who",row="alive")
```

```python
sns.relplot(data=t,x="age",y="fare",hue="class",kind="line")
```
