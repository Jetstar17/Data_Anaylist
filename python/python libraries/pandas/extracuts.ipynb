```python
import pandas as pd


marks = pd.Series([35, 48, 67, 75, 91])

grades = pd.cut(marks,bins=[0, 40, 60, 75, 100],labels=["Fail", "Pass", "Good", "Excellent"],right=False)

print(grades)

```

```python
import pandas as pd

age = pd.Series([5, 17, 25, 45, 70])

age_group = pd.cut(
    age,
    bins=[0, 18, 40, 60, 100],
    labels=["Child", "Young Adult", "Adult", "Senior"],
    right=False
)

print(age_group)
```
