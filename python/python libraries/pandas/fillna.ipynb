```python
import pandas as pd
import numpy as np

info = pd.DataFrame([[2, 7]] * 4, columns=['P', 'Q'])

print(info)

print(info.apply(np.sqrt))

print()

print(info.apply(np.sum, axis=0))

print()

print(info.apply(np.sum, axis=1))

print()

print(info)
```

```python
import pandas as pd
import numpy as np

info = pd.DataFrame({'num': np.random.randint(1, 10, 7)})    #index 7 and  btw 1 to 9

print(info)

print()

info = pd.cut(info['num'],bins=[1, 5, 10],labels=['Lows', 'Highs'],right=False)

print(info)





print(info.unique())
```

```python
import pandas as pd
import numpy as np

info = pd.DataFrame({'num': np.random.randint(1, 10, 7)})

print(info)

info['numlab'] = pd.cut(x=info['num'] ,bins=[1, 7, 10],labels=['Lows', 'Highs'],right=False)

print(info)

print(info['numlab'].unique())
```
