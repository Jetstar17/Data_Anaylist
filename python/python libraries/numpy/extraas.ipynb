```python
 import  numpy  as  np 
 import  numpy.matlib 
 # --- matlib.identity --- 
 print(np.matlib.identity(  2  ))  # Example 1: 2x2 Identity 
 print(np.matlib.identity(  5  ))  # Example 2: 5x5 Identity 
 print(np.matlib.identity(  3  , dtype=  int  ))  # Example  3: Integer type identity 
 # --- matlib.rand --- 
 print(np.matlib.rand( 3 , 3  ))  # Example 1: 3x3 Random  floats 
 print(np.matlib.rand( 1 , 5  ))  # Example 2: 1 row, 5  cols 
 print(np.matlib.rand( 2 ,  2  ) *  10  )  # Example 3: Random  values up to 10 
 # --- matlib.empty --- 
 print(np.matlib.empty((  2  ,  2  )))  # Example 1: Fast allocation 
 print(np.matlib.empty((  3  ,  1  )))  # Example 2: 3x1 empty  matrix 
```
