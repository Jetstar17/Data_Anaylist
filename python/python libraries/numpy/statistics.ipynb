```python
#amx amin
import numpy as np

a=np.array([[4,3,5,2],[8,6,7,4],[1,11,5,8]])
print(a)

print(np.amax(a))

print(np.amin(a))

print(np.amax(a,axis=1))

print(np.amin(a,axis=1))

print(np.amax(a,axis=0))

print(np.amin(a,axis=0))

print(np.amax(a,axis=1))
```

```python
print(a)
#ptp means last-1st
print(np.ptp(a,axis=1))
print(np.ptp(a,axis=0))
```

```python
b=([20,40,30,10])
print(b)

print(np.percentile(b,40))


```

```python
#med mean avg

print(a)

print(np.median(a,0))
print(np.mean(a,0))
print(np.average(a,1))
```

```python
#sort
print(a)
print(np.sort(a))
print(np.sort(a,0))
print(np.sort(a,1))
```

```python
#important
import numpy as np

d = np.dtype([("name", "S10"), ("marks", int)])

ar = np.array([("mukesh", 200), ("jay", 330)], dtype=d)

print(np.sort(ar, order="name"))
```

```python
import numpy as np
arglist=np.array([30,20,10,40])
print(arglist)

sorted= np.argsort(arglist)
print(sorted)



for i in sorted:
    print(arglist[i])
    
```
