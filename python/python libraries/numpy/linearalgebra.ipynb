```python
import numpy as np




a=np.array([[3,4],[5,6]])
b=np.array([[8,7],[9,1]])
print(np.dot(a,b))
print(np.vdot(a,b))

```

```python
a=np.array([[3,4,5,6]])
b=np.array([[8,7,9,1]])
print(np.inner(a,b))
```

```python
a=np.array([[3,4],[5,6]])
b=np.array([[8,7],[9,1]])

print(np.matmul(a,b))
```

```python
a=np.array([[3,4],[5,6]])
print(np.linalg.det(a))
```
