```python
import numpy as np
import numpy.matlib

print(np.matlib.empty((2,3)))
print(np.matlib.zeros((2,3)))
print(np.matlib.ones((2,3)))
```

```python
import numpy.matlib

# n is row, M is col, k is ind of diag
print(numpy.matlib.eye(n=3,M=3,k=0,dtype=int))
```

```python
print(numpy.matlib.identity(3,dtype=int))
print(numpy.matlib.rand(3,3))  #btw 0 and 1 onlyy
```
