```python
import numpy as np

a=np.empty((4,3),dtype=int)
print(a)
v=np.zeros((2,3),dtype=int)
print(v)
c=np.ones((3,3), dtype=int,order="C")
print(c)
c=np.ones((3,3), dtype=int)
print(c)

```

```python
a=np.arange(1,10,1,float)
print(a)
b=np.linspace(1,6,4)
print(b)
c=np.logspace(1,10,num=5,base=2,endpoint=True)
print(c)
```

```python
a=np.arange(1,13,1).reshape(3,4)
print(a)
for f in np.nditer(a):
 
 print(f,end=" ")
print()

g=a.T
print(g)
```

```python
a=np.arange(1,13,1).reshape(3,4)
print(a)

g=a.T
print(g)

for f in np.nditer(g):
 print(f,end=" ")
print()

f= g.copy(order="C")
print(f)

for m in np.nditer(f):
    print(m)
    
k=f.copy(order="F")
print(k)

for n in np.nditer(k):
 print(n)
```
