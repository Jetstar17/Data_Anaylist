```python
import numpy as np

arr = np.arange(1, 13).reshape(2, 2, 3)

print("3D Array:", arr)
print("Shape:", arr.shape)
print("Dimensions:", arr.ndim)

print("All elements:", arr.flatten())
print(arr[0,0,0])
print(arr[0,0,1])
print(arr[0,0,2])
print(arr[0,1,0])
print(arr[0,1,1])
print(arr[0,1,2])
print(arr[1,0,0])
print(arr[1,0,1])
print(arr[1,0,2])
print(arr[1,1,0])
print(arr[1,1,1])
print(arr[1,1,2])

```

```python
a = np.array([5, 2, 9, 1, 7, 6, 3, 8, 4, 10])

print("Array:", a)
print("Maximum:", a.max())
print("Minimum:", a.min())
print("Sum of elements:", a.sum())
print("Sum of 2 elements:", a[1] + a[4])

#min max sum only a.
```

```python
import numpy as np

a = np.arange(1, 7).reshape(2, 3)

print("2D Array:", a)
print("Max of each row:", a.max(axis=1))
print("Min of each column:", a.min(axis=0))
print("Sum of each row:", a.sum(axis=1))
```

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([[6, 5, 4], [3, 2, 1]])

print("a:", a)
print("b:", b)
print("a+b:", a + b)
print("a-b:", a - b)
print("a*b:", a * b)
print("a/b:", a / b)
print("Horizontal concat\n:", np.hstack((a, b)))
print("Vertical concat\n:", np.vstack((a, b)))
#np needs for stack
```
