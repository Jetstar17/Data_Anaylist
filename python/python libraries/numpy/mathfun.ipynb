```python
import numpy as np
a=np.array([10,20,30])
print(np.sin(a*np.pi/180))
print(np.cos(a*np.pi/180))
print(np.tan(a*np.pi/180))
```

```python
a=np.array([10,20,30])
sv=np.sin(a*np.pi/180)
print(sv)
cosec=np.arcsin(sv)
print(cosec)
dg=np.degrees(cosec)
print(dg)

cosv=np.cos(a*np.pi/180)
print(cosv)
sec=np.arccos(cosv)
print(sec)
dg2=np.degrees(sec)
print(dg2)


tn=np.tan(a*np.pi/180)
print(tn)
cot=np.arctan(tn)
print(cot)
dg3=np.degrees(cot)
print(dg3)
```
