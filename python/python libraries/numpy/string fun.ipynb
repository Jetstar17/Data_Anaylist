```python
import numpy as np



print(np.char.add([' hi ',' python '], [' welcome ' ' to ']))
print(np.char.multiply("hi",3))
print(np.char.center("hi",20,"*"))
print(np.char.capitalize("hi welcome"))
print(np.char.upper("hi welcome"))
print(np.char.lower("HI welcome"))
print(np.char.title("hi welcome"))
print(np.char.split("I AM JAY"))
print(np.char.splitlines("I\nAM\nJAY"))
print(np.char.strip("    I AM JAY         "))
print(np.char.join("!","jay"))

```
