```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.scatter(x,y)
```

```python
x=[5,10,15]
y=[13,26,39]
c=["red","blue","green"]
plt.scatter(x,y,color=c,marker="x")
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
x1=[5,10,15]
y1=[13,26,39]
plt.scatter(x,y,color="red",marker="o",label="Data 1")
plt.scatter(x1,y1,color="blue",marker="x",label="Data 2")
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc="upper left")
```
