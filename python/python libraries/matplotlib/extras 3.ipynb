```python

import matplotlib.pyplot as plt
from matplotlib import style
days=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
hours=[1, 2, 3, 4, 5, 6, 7]
plt.title("Hours Spent on study")
plt.xlabel("Days of the week")
plt.ylabel("Hours")
plt.plot(days, hours,color='green', marker='*')


```

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
quantity = [10, 15, 7, 12]
colors=['red', 'yellow', 'pink', 'brown']
plt.bar(fruits, quantity, color=colors)
plt.title("Fruit inventory")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.grid(axis='y', linestyle='-', alpha=1) #alpha controls the transparency of the grid lines range 0 to 1

```

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
quantity = [10, 15, 7, 12]
colors=['red', 'yellow', 'pink', 'brown']
plt.bar(fruits, quantity, color=colors,hatch='//') #hatch adds pattern to the bars
plt.title("Fruit inventory")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.grid(axis='y', linestyle='-', alpha=1) #alpha controls the transparency of the grid lines range 0 to 1

```

```python
toppings = ['Pepperoni', 'Mushrooms', 'Onions', 'Sausage', 'Bacon']
votes = [50, 30, 20, 40, 60]
explode = (0.1, 0, 0, 0, 0) #explode the first slice
plt.pie(votes, labels=toppings, autopct='%1.1f%%', shadow=True, explode=explode)
plt.title("pizza party planning")

```
