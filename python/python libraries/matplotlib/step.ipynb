```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
plt.step(x, y)

```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
plt.step(x, y, where='post') #look after the step   
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
plt.step(x, y, where='pre') #look before the step
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
plt.step(x, y, where='mid')   #look mid point of the step
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
plt.step(x, y,"g*", where='post')    #green color and star marker only visible at the step point not at the step line
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
plt.step(x, y,"cs", where='post') #cyan color and square marker only visible at the step point not at the step line
plt.xlim(1,9) #x limit is set to 1 to 9 to show the step line after the last point at x=5
```
