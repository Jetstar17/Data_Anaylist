```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.1, 2*np.pi,41)
y = np.exp(np.sin(x))
plt.stem(x, y)
```

```python
x=np.linspace(0.1, 2*np.pi,41)
y=np.exp(np.sin(x))
plt.stem(x, y,bottom=1.1)
```

```python
x=np.linspace(0.1, 2*np.pi,41)
y=np.exp(np.sin(x))
markerline, stemlines, base_line = plt.stem(x, y,bottom=1.1,linefmt='red', markerfmt='blue', basefmt='gold')
markerline.set_markerfacecolor('none')  #not direct parameter of stem must returned to set the marker face color
```
