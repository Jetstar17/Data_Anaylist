```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
x=[1, 2, 3]
y=[7, 8, 9]
x1=[4, 5, 6]
y1=[10, 11, 12]
plt.bar(x, y, color="r", align="center", width=0.5)
plt.bar(x1, y1, color="g", align="center", width=0.5)
```

```python
import matplotlib.pyplot as plt
import numpy as np
style.use("ggplot")
countries = ["India", "USA", "China", "Russia", "UK"]
gold = np.array([10, 20, 30, 40, 50])
silver = np.array([15, 25, 35, 45, 55])
bronze = np.array([20, 30, 40, 50, 60])

plt.bar(countries, bronze, color="#CD7F32", label="Bronze")
plt.bar(countries, silver, bottom=bronze, color="silver", label="Silver")
plt.bar(countries, gold, bottom=bronze+silver, color="gold", label="Gold")

plt.xlabel("Countries")
plt.ylabel("Medals")
plt.title("olympics 2026")
plt.legend(loc="upper left")


```
