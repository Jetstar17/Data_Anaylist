```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
from matplotlib import style

marks=[20,30,40]
names=["jay","krish","kartik"]
explode=[0.1,0.1,0.1]

plt.pie(marks,labels=names,explode=explode,shadow=True,autopct="%1.1f%%")
```
