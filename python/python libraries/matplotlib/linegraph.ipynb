```python
import matplotlib.pyplot as plt

a=[1, 2, 3, 4, 5]
b=[6, 7, 8, 9, 10]
plt.plot(a, b)
```

```python
X=[1, 2, 3, 4, 5]
Y=[6, 7, 8, 9, 10]  
c="r"
plt.plot(X, Y, color=c)
```

```python
a=[8,9,12,7,10]
b=[6, 7, 8, 9, 10]
c="g"
plt.plot(a, b, color=c, marker="o", linestyle="dashed", linewidth=2, markersize=12)
```

```python
a=[9,8,5,6,7]
b=[2,6,1,8,9]
c="b"
d="g"
e="r"
plt.plot(a, b, color=c, marker="o", linestyle="dashed", linewidth=2, markersize=12)
plt.plot(a,color=d, marker="*", linestyle="dashed", linewidth=2, markersize=12)
plt.plot(b,color=e, marker="+", linestyle="dashed", linewidth=2, markersize=12)
```
