```python

import matplotlib.pyplot as plt 
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'] 
steps = [3000, 4500, 5000, 2000, 7000, 9000, 8500] 
plt.plot(days, steps, color='blue', marker='o', linestyle='-') 
plt.title("Fitness Tracker: Steps per Day") 
plt.xlabel("Day of the Week") 
plt.ylabel("Number of Steps") 
plt.grid(True) 
```

```python
import matplotlib.pyplot as plt 
weeks = [1, 2, 3, 4] 
plant1_height = [2, 5, 8, 12] 
plant2_height = [1, 3, 6, 10] 
plt.plot(weeks, plant1_height, label='Plant A', color='green') 
plt.plot(weeks, plant2_height, label='Plant B', color='red', linestyle='--') 
plt.title("Plant Growth Over Time") 
plt.legend() # Shows which color belongs to which plant
```

```python

import matplotlib.pyplot as plt 
temp = [20, 25, 30, 35, 40, 45] 
sales = [100, 150, 250, 400, 600, 800] 
plt.scatter(temp, sales, color='orange', marker='s') # 's' is for square marker 
plt.title("Temperature vs Ice Cream Sales") 
plt.xlabel("Temperature (°C)") 
plt.ylabel("Sales ($)") 
```

```python
import matplotlib.pyplot as plt 
hours = [1, 2, 3, 4, 5, 6, 7, 8] 
scores = [45, 50, 55, 70, 75, 88, 92, 98] 
plt.scatter(hours, scores, color='purple', marker='x') 
plt.title("Study Time vs Exam Results")
```

```python
import matplotlib.pyplot as plt 
products = ['Laptop', 'Mobile', 'Tablet', 'Watch'] 
units_sold = [50, 120, 40, 90] 
plt.bar(products, units_sold, color=['navy', 'blue', 'skyblue', 'cyan']) 
plt.title("Electronic Store Sales") 
plt.ylabel("Units Sold") 
```

```python
import matplotlib.pyplot as plt 
movies = ['Inception', 'Avatar', 'Titanic', 'Joker'] 
ratings = [8.8, 7.8, 7.9, 8.4] 
plt.barh(movies, ratings, color='teal') # barh creates horizontal bars 
plt.title("Movie Ratings Comparison") 
plt.xlabel("IMDb Rating")
```

```python
import matplotlib.pyplot as plt 
categories = ['Rent', 'Food', 'Travel', 'Savings'] 
amount = [40, 25, 15, 20] 
plt.pie(amount, labels=categories, autopct='%1.1f%%', shadow=True) 
plt.title("Monthly Expense Distribution") 
plt.show() 
```

```python
import matplotlib.pyplot as plt 
languages = ['Python', 'Java', 'C++', 'JavaScript'] 
popularity = [45, 20, 15, 20] 
explode = [0.1, 0, 0, 0] # This pops out the first slice (Python) 
plt.pie(popularity, labels=languages, explode=explode, autopct='%1.1f%%') 
plt.title("Language Popularity")
```
