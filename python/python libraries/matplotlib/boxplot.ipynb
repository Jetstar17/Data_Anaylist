```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
data=np.random.normal(100,20,200) 

figure=plt.figure(figsize=(5,2))  #figure size is set to 5 inch width and 2 inch height
plt.boxplot(data)
```
