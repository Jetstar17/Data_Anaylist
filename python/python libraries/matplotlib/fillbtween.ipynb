```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-5,5,0.1)
y1=-3*x**2+5*x+10
y2=3*x**2-5*x-10
plt.plot(x,y1,x,y2)
plt.fill_between(x,y1,y2,where=(y1>y2),color='red',alpha=0.5)
plt.fill_between(x,y1,y2,where=(y1<y2),color='blue',alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area between two curves')
plt.legend(['y1=-3x^2+5x+10','y2=3x^2-5x-10'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2, 0.01)

y1 = np.sin(2 * np.pi * x) 

y2 = 0.8 * np.sin(4 * np.pi * x)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, sharex=True, figsize=(6,6))

# 1️⃣ Fill between y1 and 0
ax1.plot(x, y1, color='black')
ax1.fill_between(x, y1, 0, facecolor='green')
ax1.set_title("Fill Between y1 and 0")

# 2️⃣ Fill between y1 and 1
ax2.plot(x, y1, color='black')
ax2.fill_between(x, y1, 1, facecolor='green')
ax2.set_title("Fill Between y1 and 1")

# 3️⃣ Fill between y1 and y2
ax3.plot(x, y1, color='black')
ax3.plot(x, y2, color='red')
ax3.fill_between(x, y1, y2, facecolor='green')
ax3.set_title("Fill Between y1 and y2")

# 4️⃣ Fill between y1 and y2 with condition
ax4.plot(x, y1, color='black')
ax4.plot(x, y2, color='red')
ax4.fill_between(x, y1, y2, where=(y2 <= y1),
                 facecolor='green')
ax4.set_title("Fill Between y1 and y2 (y2 <= y1)")

plt.tight_layout()
plt.show()
```
