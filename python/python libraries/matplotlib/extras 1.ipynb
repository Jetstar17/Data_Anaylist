```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
bench_press = [60, 65, 70, 75, 80, 85]

plt.plot(months, bench_press, marker='o', linestyle='--')
plt.title("Bench Press Progress (kg)")
plt.xlabel("Month")
plt.ylabel("Weight (kg)")
plt.grid(True)  #same as style.use('ggplot')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create figure
plt.figure(figsize=(12, 8))

# ---------------- 1️⃣ Line Plot ----------------
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
bench_press = [60, 65, 70, 75, 80, 85]

plt.subplot(2, 2, 1)
plt.plot(months, bench_press, marker='o')
plt.title("Bench Press Progress")
plt.xlabel("Month")
plt.ylabel("Weight (kg)")
plt.grid(True)

# ---------------- 2️⃣ Bar Chart ----------------
sources = ["Paneer", "Dal", "Milk", "Soy", "Peanuts"]
protein = [18, 9, 8, 15, 7]

plt.subplot(2, 2, 2)
plt.bar(sources, protein)
plt.title("Protein per 100g")
plt.xlabel("Source")
plt.ylabel("Protein (g)")

# ---------------- 3️⃣ Pie Chart ----------------
activities = ["Study", "Gym", "Sleep", "Coding", "Other"]
hours = [5, 1.5, 7, 4, 6.5]

plt.subplot(2, 2, 3)
plt.pie(hours, labels=activities, autopct="%1.1f%%")
plt.title("Daily Time Distribution")

# ---------------- 4️⃣ Scatter Plot ----------------
height = [160, 165, 170, 175, 180, 185]
weight = [55, 60, 68, 72, 80, 85]

plt.subplot(2, 2, 4)
plt.scatter(height, weight)
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

subjects = ["Math", "Physics", "Chemistry", "Coding"]
marks = [85, 78, 88, 95]

plt.bar(subjects, marks, color=["skyblue", "orange", "green", "purple"])

plt.title("Semester Marks", fontsize=14)
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Marks", fontsize=12)
plt.ylim(0, 100)

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
bench = [60, 65, 70, 75, 80, 85]
squat = [80, 90, 100, 110, 120, 130]

plt.plot(months, bench, marker='o', label="Bench Press")
plt.plot(months, squat, marker='s', label="Squat")

plt.title("Strength Progress Over 6 Months")
plt.xlabel("Month")
plt.ylabel("Weight (kg)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
steps = np.random.randint(3000, 12000, 30)

plt.hist(steps, bins=6, edgecolor="black")

plt.title("Daily Steps Distribution (30 Days)")
plt.xlabel("Steps")
plt.ylabel("Frequency")
plt.show()
```
